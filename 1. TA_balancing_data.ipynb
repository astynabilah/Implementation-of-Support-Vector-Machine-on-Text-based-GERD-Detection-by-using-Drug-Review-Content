```
import numpy as np
import pandas as pd
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
downloaded = drive.CreateFile({'id':"1UCG-4hHPotw7_75PcQMpAz18C4DS0TVO"})
downloaded.GetContentFile('train.csv') 
train = pd.read_csv('train.csv') 
downloaded = drive.CreateFile({'id':"1s0l8ucZb1MT2fhumU7NdzVl3etImX-lk"})
downloaded.GetContentFile('test.csv') 
test = pd.read_csv('test.csv') 
```

```
train = train[['condition','review']]
train = train.drop_duplicates(subset='review', keep='first')
test = test[['condition','review']]
test = test.drop_duplicates(subset='review', keep='first')
```

```
train['condition'] = (train['condition'] == 'GERD').astype(int)
test['condition'] = (test['condition'] == 'GERD').astype(int)
```

```
train
```

```
test
```

```
gerd_amt_train = (train.condition==1).sum()
gerd_amt_test = (test.condition==1).sum()
gerd_amt_train, gerd_amt_test
```

```
train_not_gerd = train.loc[train['condition']==0]
train_gerd = train[train['condition']==1]
test_not_gerd = test.loc[test['condition']==0]
test_gerd = test[test['condition']==1]
```

```
train_gerd
```

```
train_non_gerd_selected = train_not_gerd.sample(n=gerd_amt_train)
test_non_gerd_selected = test_not_gerd.sample(n=gerd_amt_test)
```

```
train_non_gerd_selected
```

```
test_non_gerd_selected
```

```
train_all = train_non_gerd_selected.append(train_gerd)
train_all = train_all.sort_index()
test_all = test_non_gerd_selected.append(test_gerd)
test_all = test_all.sort_index()
```

```
test_all
```

```
from google.colab import files

train_all.to_csv('train_balanced.csv')
files.download('train_balanced.csv')
test_all.to_csv('test_balanced.csv')
files.download('test_balanced.csv')
```
